```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.cos(x)
```

```python
xx = np.linspace(-2,2,100)
```

```python
plt.figure()
plt.plot(xx,xx)
plt.plot(xx,f(xx))
plt.xlabel(r"$x_n$")
plt.ylabel(r"$x_{n+1}$")
plt.show()
```

```python
x0 = -0.5
```

```python
f(f(f(f(f(f(x0))))))
```

```python
xn_arr = []
xn_arr.append(x0)

for i in range(100):
    xn_arr.append(f(xn_arr[-1]))
```

```python
plt.figure()
plt.plot(xn_arr)
plt.ylabel(r"$x_n$")
plt.xlabel(r"$n$")
plt.show()
```
